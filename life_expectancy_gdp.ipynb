```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# First, I will load the data 

life = pd.read_csv('all_data.csv')
```

```python
# print first 10 rows and observe columns
print(life.head(10))
```

```python
# perform summary statistics to understand data better
life.describe()
```

```python
# How many countries are involved and what are the countries?
print(life.Country.nunique())
print(life.Country.unique())
```

```python
# average life expenctancy for each country between the year 2000 and 2015

average_expectancy = life['Life expectancy at birth (years)'].groupby(life.Country).mean()
print(average_expectancy)
```

```python
# average GDP for each country between the year 2000 and 2015

average_gdp = life['GDP'].groupby(life.Country).mean()
print(average_gdp)
```

```python
# Has GDP increased over time?

sns.lineplot(life.Year, life.GDP, hue = life.Country)

plt.xticks(rotation = 40)
plt.xlabel('Year')
plt.ylabel('($)GDP')
plt.title('GDP Trend Overtime')
plt.show()
```

```python
# Life expectancy trend over time
sns.lineplot(life.Year, life['Life expectancy at birth (years)'], hue = life.Country)

plt.xticks(rotation = 40)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy Overtime Across Six Countries')
plt.show()
```

```python
# Relationship between GDP and Life expectancy
sns.scatterplot(y = life.GDP, x= life['Life expectancy at birth (years)'], hue = life.Country)
plt.show()
```
